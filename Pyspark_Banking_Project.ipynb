```python
import pyspark
from pyspark.sql import SparkSession
from pyspark.storagelevel import StorageLevel
from pyspark.sql.functions import *
```

```python
null_count = df.select([sum(when(col(c).isNull(),1).otherwise(0)).alias(c) for c in df.columns])
# Convert to dict
null_dict = null_count.collect()[0].asDict()
# Filter columns with nulls
columns_with_nulls = [c for c, v in null_dict.items() if v > 0]
```
